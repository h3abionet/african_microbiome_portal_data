```python
def rearrange_feature(x):
    try:
        x = x.split('//')
        return int(x[0])
        # return f"{x[1]}, {x[0]} years old"
    except:
        return x



import pandas as pd
import numpy as np
data = pd.read_csv("../Originals/Fragmented/PRJNA559354.csv")
data['ELO']=''
data['PARTICIPANT FEATURES']= data['PARTICIPANT FEATURES'].apply(rearrange_feature)

data['LAT LON']="-33.918861,18.423300"


print(set(data['PARTICIPANT FEATURES']))

from replacements import replacement_func
cols = ["STUDY TITLE","ASSAY TYPE","TECHNOLOGY","COUNTRY","STUDY DESIGN","BODY SITE","PARTICIPANT FEATURES","ETHNICITY","ELO", "LIBRARY LAYOUT","BODY SITE", "URBANZATION", "SAMPLE TYPE","REGION","DIET", "CITYVILLAGE"] 
data[cols] = data[cols].applymap(lambda x: replacement_func(x) if type(x)==str else x).values
for col in data.columns:
    data.loc[pd.isna(data[col]), col] = np.nan
    data.loc[data[col]=="-", col] = np.nan
    try:
        if col not in cols:
            data[col] = data[col].apply(lambda x: x.upper().strip())
        data.loc[data[col]=="", col] = np.nan
    except:
        pass
    
    print(col)
    print("-------------")
    print(set(data[col]), len(set(data[col])))
    
    print("=========================================================")

    
data.to_csv("../Corrected/PRJNA559354.csv")
```
